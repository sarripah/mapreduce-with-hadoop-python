```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df.head()
```

```python
# Menghapus baris dengan nilai "1 Season" atau "2 Seasons" di kolom "duration"
df = df[~df['duration'].isin(['1 Season', '2 Seasons'])]
```

```python
# Menghapus kata "min" dan menyimpan hanya angka di kolom "duration"
df['duration'] = df['duration'].str.extract('(\d+)').astype(float)
```

```python
#show unique values
df['country'].unique()
```

```python
df['country'].isna().sum()
```

```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['country'] = df['country'].astype(str)
df['country'] = df['country'].apply(lambda x : x.split(', ')[0])
```

```python
#show unique values
df['rating'].unique()
```

```python
df['rating'] = df['rating'].replace({'74 min' : np.nan, '84 min' : np.nan, '66 min': np.nan
                                    , 'TV-Y7-FV' : 'TV-Y7'})
df['rating'].unique()
```

```python
df['director'].isnull().sum()
```

```python
df['director'].value_counts()
```

```python
#compute the percentage of null values 
null_percent = df.isnull().sum() * 100 / df.shape[0]
null_percent.round(2).sort_values(ascending = False)
```

```python
#fill null values with new category
df['director'].fillna('Unkown', inplace= True)
df['cast'].fillna('Unkown', inplace= True)
```

```python
#fill null values with mode
mode_im = ['date_added','rating','duration']
for i in mode_im:
    df[i] = df[i].fillna(df[i].mode()[0])
```

```python
#drop useless columns
df.drop(['date_added','description'],axis=1, inplace= True)
df.isnull().sum()
```

```python
# Menyimpan DataFrame setelah pemrosesan ke file CSV baru
df.to_csv('netflix_clean.csv', index=False)
```
